```python
#backtracking system
def predict(train,test,predictors,model):
  model.fit(train[predictors],train["Target"])
  preds=model.predict(test[predictors])
  preds=pd.Series(preds,index=test.index,name="Predictions")
  combined=pd.concat([test["Target"],preds],axis=1)
  return combined
```

```python

def backtest(data,model,predictors,start=2500,step=250):
  all_predictions=[]

  for i in range(start,data.shape[0],step):
    train=data.iloc[0:i].copy()
    test=data.iloc[i:(i+step)].copy()
    predictions=predict(train,test,predictors,model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```

```python
#applying weights
# Weighted predict function
def predict(train, test, predictors, model):
    # Create weights: older samples get lower, recent samples get higher weight
    weights = np.linspace(0.1, 1.0, len(train))

    # Fit the model with weights
    model.fit(train[predictors], train["Target"], sample_weight=weights)

    # Predict probabilities
    preds = model.predict_proba(test[predictors])[:,1]

    # Apply decision threshold
    preds[preds >= 0.6] = 1
    preds[preds < 0.6] = 0

    # Format predictions
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)

    return combined
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score

# Weighted predict function
def predict(train, test, predictors, model):
    weights = np.linspace(0.1, 1.0, len(train))
    model.fit(train[predictors], train["Target"], sample_weight=weights)
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= 0.6] = 1
    preds[preds < 0.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined

# Backtesting loop
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i + step)].copy()
        print(f"Training on rows 0 to {i}")
        print(f"Testing on rows {i} to {i + step}")
        print(f"Train shape: {train.shape}, Test shape: {test.shape}")
        predictions = predict(train, test, predictors, model)
        if predictions is not None and not predictions.empty:
            all_predictions.append(predictions)
        else:
            print(f"No predictions returned for range {i}-{i + step}")
    if all_predictions:
        return pd.concat(all_predictions)
    else:
        raise ValueError("No valid predictions were generated during backtesting.")

# Instantiate and train the model
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)

# Run your backtest
predictions = backtest(sp500, model, new_predictors)

# Evaluation
print("Prediction distribution:")
print(predictions['Predictions'].value_counts())

print("Precision Score:")
print(precision_score(predictions["Target"], predictions["Predictions"]))
```

```python
#SVM
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
def predict(train, test, predictors, model):
    scaler = StandardScaler()

    # Fit scaler on training data and transform both train and test
    train_scaled = scaler.fit_transform(train[predictors])
    test_scaled = scaler.transform(test[predictors])

    model.fit(train_scaled, train["Target"])
    preds = model.predict(test_scaled)

    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:i + step].copy()
        predictions = predict(train, test, predictors, model)

        if predictions is not None and not predictions.empty:
            all_predictions.append(predictions)

    if all_predictions:
        return pd.concat(all_predictions)
    else:
        raise ValueError("No valid predictions were generated during backtesting.")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from prophet import Prophet
import matplotlib.pyplot as plt

# Step 1: Load the S&P 500 data
sp500 = yf.Ticker("^GSPC").history(period="max")

# Step 2: Prepare the DataFrame for Prophet
df = sp500.reset_index()[["Date", "Close"]].copy()
df.rename(columns={"Date": "ds", "Close": "y"}, inplace=True)

# Step 3: Remove timezone info
df["ds"] = df["ds"].dt.tz_localize(None)

# Step 4: Apply log transformation to stabilize variance
df["y"] = np.log(df["y"])

# Step 5: Initialize and fit the Prophet model
model = Prophet(daily_seasonality=True)
model.fit(df)

# Step 6: Create future dataframe for next 365 days
future = model.make_future_dataframe(periods=365)

# Step 7: Predict future values
forecast = model.predict(future)

# Step 8: Exponentiate the predicted values to return to original scale
forecast["yhat"] = np.exp(forecast["yhat"])
forecast["yhat_lower"] = np.exp(forecast["yhat_lower"])
forecast["yhat_upper"] = np.exp(forecast["yhat_upper"])

# Step 9: Plot the forecast
fig = model.plot(forecast)
plt.title("S&P 500 Forecast (Using Prophet with Log Transformation)")
plt.xlabel("Date")
plt.ylabel("S&P 500 Close Price")
plt.show()

```

```python
# Create a binary target for the actual S&P 500 data
df["Target"] = (df["y"].diff().shift(-1) > 0).astype(int)  # 1 if the next day's price is higher, else 0
```

```python
# For predicted values: create the same binary target
forecast["Predictions"] = (forecast["yhat"].diff().shift(-1) > 0).astype(int)
```

```python
from sklearn.metrics import precision_score

# Align forecast to match original df's date range
predictions = forecast[forecast['ds'] <= df['ds'].max()].copy()
predictions = predictions.set_index('ds')

# Align original Target values
actual = df.set_index('ds')

# Ensure both series are aligned
predictions = predictions.loc[actual.index]

# Calculate precision score
precision = precision_score(actual["Target"], predictions["Predictions"])
print(f"Precision Score: {precision:.4f}")
```
